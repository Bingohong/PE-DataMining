```python
label = data.ROP
dataset = data.drop(["ROP"],axis=1)
x_tmp,x_test,y_tmp,y_test = train_test_split(dataset,label,test_size=0.15)
x_train,x_val,y_train,y_val = train_test_split(x_tmp,y_tmp,test_size=0.18)
print("x_tmp shape: ",x_tmp.shape)
print("x_train shape: ",x_train.shape)
print("x_val shape: ",x_val.shape)
print("x_test shape: ",x_test.shape)
```

#### 算法&度量指标

```python
from sklearn.metrics import mean_squared_log_error,mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.svm import SVR

# 不同数量的训练数据所得的验证和训练分数
from sklearn.model_selection import learning_curve 
# 单个超参数与训练分数和验证分数的关系图
from sklearn.model_selection import validation_curve
# 超参数搜索
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_validate, cross_val_score

from sklearn.ensemble import partial_dependence
```

```python
def get_model(estimator,parameters,X_train=x_tmp,y_train=y_tmp,scoring="neg_mean_squared_log_error",n_jobs=2,cv=5,verbose=0):
    model = GridSearchCV(estimator,param_grid=parameters,scoring=scoring,return_train_score=True,cv=5,verbose=verbose)
    model = model.fit(X_train,y_train)
    return model
```

##### 样本学习曲线

```python
def plot_learning_curve(estimator,title,X=x_tmp,y=y_tmp,ylim=None,cv=5,
                       n_jobs=2,train_sizes=np.linspace(.1,1.0,5),scoring="neg_mean_squared_error"):
    plt.figure(figsize=(6,4))
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel(scoring)
    train_sizes,train_scores,test_scores=learning_curve(estimator,X,y,cv=cv,scoring=scoring,
                                                       n_jobs=n_jobs,train_sizes=train_sizes)
    train_scores_mean=np.mean(train_scores,axis=1)
    train_scores_std=np.std(train_scores,axis=1)
    test_scores_mean=np.mean(test_scores,axis=1)
    test_scores_std=np.std(test_scores,axis=1)
    
    plt.grid()
    
    plt.fill_between(train_sizes,train_scores_mean-train_scores_std,
                    train_scores_mean+train_scores_std,alpha=0.2,color="r")
    
    plt.fill_between(train_sizes,test_scores_mean-test_scores_std,
                    test_scores_mean+test_scores_std,alpha=0.1,color="g")
    
    plt.plot(train_sizes,train_scores_mean,"o-",color="r",label="Training score")
    plt.plot(train_sizes,test_scores_mean,"o-",color="g",label="Cross_validation score")
    
    plt.legend(loc="best")
    
    return plt
```

##### 交叉验证训练

```python
def cv_fit(estimator,x=x_tmp,y=y_tmp,cv=5,scoring="neg_mean_squared_error"):
    result = cross_validate(estimator,X=x,y=y_tmp,cv=5,scoring=scoring)
    train_score = result["train_score"]
    test_score = result["test_score"]
    train_score_mean = np.mean(train_score)
    test_score_mean = np.mean(test_score)
    print("train_score mean: ",train_score_mean)
    print("test_score mean: ",test_score_mean)
    fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(6,3),sharey=True)
    axes[0].scatter(x=range(len(train_score)),y=train_score)
    axes[0].axhline(train_score_mean,color="r")
    axes[0].set_title("train_score")
    axes[1].scatter(x=range(len(test_score)),y=test_score)
    axes[1].axhline(test_score_mean,color="r")
    axes[1].set_title("test_score")
    return result
```

```python
def model_predict(estimator,x=x_test,y=y_test):
    pred = estimator.predict(x)
    error = mean_squared_error(y_test,pred)
    print("the error_log is %.4f:"%(error))
    res_log = pred - y_test
    res = np.exp(pred) - np.exp(y_test)
    fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(10,4))
    axes[0].scatter(range(len(res_log)),res_log)
    axes[0].set_title("res_log")
    axes[1].scatter(range(len(res)),res)
    axes[1].set_title("res")
    return pred
```

##### RandomForest方法

```python
rf = RandomForestRegressor(n_estimators=20,criterion="mse",min_samples_leaf=5,min_impurity_decrease=0.02,max_depth=7,verbose=0)
#cross_validate(rf,X=x_tmp,y=y_tmp,scoring="neg_mean_squared_error",cv=5)
#rfer = rf.fit(x_train,y_train)
result = cv_fit(rf)
```
